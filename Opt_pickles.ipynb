```python
import pandas as pd
import pickle
import matplotlib.pyplot as plt

def plot_boxplot(sample_percentage, x_axis_variable):
    # Load the DataFrame from pickle
    with open(f'study_results_20.pkl', 'rb') as f:
        df = pickle.load(f)

    # Now you can work with the DataFrame 'df'
    #print(df.columns)
    #df_params = pd.json_normalize(df['params'])

    # Concatenate the exploded params DataFrame with the original DataFrame
    #df_exploded = pd.concat([df.drop(columns=['params']), df_params], axis=1)
    #df_attrs = pd.json_normalize(df_exploded['user_attrs'])

    # Concatenate the exploded params DataFrame with the original DataFrame
    #df_exploded = pd.concat([df_exploded.drop(columns=['user_attrs']), df_attrs], axis=1)
    df_exploded = df
    df_exploded['FMgrace'] = round(df_exploded['FMgrace'] / df_exploded['train_packets'], 2)
    
    df_exploded['FPR'] = df_exploded['anomaly_count'] / df_exploded['test_convs']
    min_fpr_index = df_exploded['FPR'].idxmin()
    
    print(f"{sample_percentage} percent")
    print(df_exploded.loc[min_fpr_index, ['train_packets', 'FMgrace', 'numAE', 'learning_rate', 'hidden_ratio', 'training_error', 'testing_error', 'FPR', 'threshold']])
    plt.figure(figsize=(10, 6))  # Adjust size if needed
    df_exploded.boxplot(column='FPR', by=x_axis_variable, grid=False)
    plt.title(f'Boxplot of Testing Error by {x_axis_variable} ({sample_percentage} Percent of Training Sample)')
    plt.xlabel(x_axis_variable)
    plt.ylabel('Testing Error')
    plt.suptitle('')  # Remove default title
    plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
    plt.tight_layout()
    plt.show()
```

```python
plot_boxplot(10, 'learning_rate')
plot_boxplot(15, 'learning_rate')
plot_boxplot(20, 'learning_rate')
plot_boxplot(25, 'learning_rate')
plot_boxplot(50, 'learning_rate')
```

```python
plot_boxplot(10, 'numAE')
plot_boxplot(15, 'numAE')
plot_boxplot(20, 'numAE')
plot_boxplot(25, 'numAE')
plot_boxplot(50, 'numAE')
```
